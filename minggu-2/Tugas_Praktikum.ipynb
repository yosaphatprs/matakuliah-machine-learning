Nomor 1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
import pandas as pd

dpath = "data/wbc.csv"
df = pd.read_csv(dpath)
df.head(20) # Menampilkan data terlebih dahulu untuk mengetahui isi dari data tersebut, supaya dapat memisahkan antara variable yang dapat digunakan dan variabel yang tidak dapat digunakan.
```

```python
df.info() # Menampilkan informasi dari data tersebut, untuk mengetahui tipe data dari setiap variabel.
```

```python
df.isnull().sum() # Melakukan pengecekan apakah terdapat kolom yang perlu melakukan proses imputasi.
```

```python
# Slice Data yang diperlukan
df = df.drop('Unnamed: 32', axis=1) # Menghapus kolom yang tidak diperlukan yaitu Unnamed: 32 sebagai upaya untuk slicing data
df.head() # Menampilkan 5 data teratas
```

Nomor 2. Lakukan proses encoding pada kolom "diagnosis".

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # membuat objek dari LabelEncoder
df['diagnosis'] = le.fit_transform(df['diagnosis']) # Melakukan proses encoding pada variabel diagnosis
df.head() # Menampilkan 5 data teratas
```

Nomor 3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler() # membuat objek dari StandardScaler
columns_to_standardize = df.select_dtypes(include=['float64']).columns # Memilih kolom yang akan distandarisasi, yaitu kolom yang memiliki nilai numerik float64
df[columns_to_standardize] = std.fit_transform(df[columns_to_standardize]) # Melakukan proses standarisasi pada kolom yang telah dipilih
df.head() # Menampilkan 5 data teratas
```

Nomor 4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20

```python
# Split data
from sklearn.model_selection import train_test_split # Import library train_test_split dari sklearn.model_selection

df.head() # Menampilkan 5 data teratas

# Split data training dan lainnya
# data lainnya, akan kita split lagi menjadi validasi dan testing.
# Rasio yang akan kita gunakan adalah 8:1:1
df_train, df_unseen = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis']) # Split data training dan lainnya dengan rasio 8:2.

# Split lagi antara validasi dan testing
df_val, df_test = train_test_split(df_unseen, test_size=0.5, random_state=0, stratify=df_unseen['diagnosis']) # Split data lainnya (subset df_unseen) menjadi validasi dan testing dengan rasio 1:1

# Cek masing-masing ukuran data

print(f'Jumlah label data asli:\n{df.diagnosis.value_counts()}') # Menampilkan jumlah label data asli
print(f'\nJumlah label data train:\n{df_train.diagnosis.value_counts()}') # Menampilkan jumlah label data train
print(f'\nJumlah label data val:\n{df_val.diagnosis.value_counts()}') # Menampilkan jumlah label data validasi
print(f'\nJumlah label data test:\n{df_test.diagnosis.value_counts()}') # Menampilkan jumlah label data testing
```
