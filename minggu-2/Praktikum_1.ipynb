# Langkah 1 - Load Data

```python
import pandas as pd # Melakukan Import Library Pandas

dpath = 'data/Titanic-Dataset.csv' # Menentukan Path Dataset Titanic
df = pd.read_csv(dpath) # Membaca Dataset Titanic dan Menyimpannya ke dalam DataFrame Pandas
df.head() # Menampilkan 5 Data Teratas karena tidak ada argumen di dalam fungsi head(), jika fungsi head() terdapat argumen berupa angka maka akan menampilkan data sebanyak angka yang ada dalam argumen
```

# Langkah 2 - Pengecekan Data

```python
df.info() # Menampilkan Informasi Data Titanic yang ada dalam DataFrame Pandas, informasi tersebut berisi seperti jumlah data setiap kolom, tipe data setiap kolom, jumlah kolom, dan jumlah baris yang ada dalam dataset.
```

```python
df.isnull().sum() # Menampilkan jumlah data yang kosong atau hilang pada setiap kolom
```

Berdasarkan hasil diatas, ditemukan bahwa terdapat 177 data yang hilang pada kolom Age, 687 data yang hilang pada kolom Cabin, dan 2 data yang hilang pada kolom Embarked.

# Langkah 3 - Imputasi

```python
# Age - mean
df['Age'].fillna(value=df['Age'].mean(), inplace=True) # Mengisi data yang kosong atau hilang pada kolom Age dengan nilai rata-rata dari kolom Age. Hal ini dikarenakan kolom Age adalah data nominal yang ditunjukkan oleh tipe data float64, sehingga imputasi yang tepat adalah menggunakan rata-rata/mean.

# Cabin - "DECK"
df['Cabin'].fillna(value="DECK", inplace=True) # Mengisi data yang kosong atau hilang pada kolom Cabin dengan nilai "DECK". Hal ini dikarenakan kita berasumsi bahwa penumpang yang tidak memiliki nomor cabin tinggal di dek kapal.

# Embarked - modus
df['Embarked'].fillna(value=df['Embarked'].mode, inplace=True) # Mengisi data yang kosong atau hilang pada kolom Embarked dengan nilai modus dari kolom Embarked. Hal ini dikarenakan kolom Embarked adalah data nominal berdasarkan nilai pada kolom Embarked, sehingga imputasi yang tepat adalah menggunakan modus.
```
